## Writing Python code


In python, we use variables and functions to perform tasks. 

### Variables

A variable is something that indicates a number, word, sentence, or in a work a variable can also be an atom or a molecule or a nanoparticle. A variable can have any name you want to give it, as long as it does not start with a number. Some examples of variables are shown below:

```python
a_variable = 1
another_variable = 'hi'
a_3rd_variable = 'This is a third (3rd) variable'
this_is_another_number = 3.14
```

You can read more about variables in https://www.w3schools.com/python/python_variables.asp

### Functions

A function is a method that will perform some algorithm. Methods are written as `a_method(variable1,variable2,variable3)`, where `a_method` is the method and `variable1,variable2,variable3` are the input variables that are required by `a_method`. Some functions may need just one input variable, some many need many variables, and other may not need any input variables. Some examples of functions are given below:

```python
def double_function(my_number):
    answer = my_number * 2
    return answer

def adding_function(number1,number2):
    return number1 + number2

def subtract_function(number1,number2):
    return number1 - number2

def multiply_function(number1,number2):
    return number1 * number2
```

These can be run as shown below:

```python
answer_is = double_function(3)
print(answer_is)
print(adding_function(1,1))
print(adding_function(2,3))
print(subtract_function(5,4))
big_number = multiply_function(adding_function(2,3),5)
print(big_number)
```

Some functions are associated of certain variables. These are written as `a_variable.associated_method(input_variable1,input_variable2)`, where `a_variable` is a variable that contains `associated_method`. Some examples are:

```python
print("This sentence will have all e's removed".replace('e',''))
a_list = [1,2,3,4,5]
a_list.remove(2)
print(a_list)
a_list.append('Hi')
print(a_list)
```

Note: It is important that you run each code box in order like a book, as the previous code boxes may be required for the next code box. 

You can read more about functions in https://www.w3schools.com/python/python_functions.asp
