# 2.4: How are these Faces related to Energetics and Stability?

The types and amounts of square and triangle surfaces that we will see in a nanoparticle are often based on their energy; the lower the energy of a surface type, the more likely we will see that surface in a nanoparticle. Often, those surface types with lower surface areas are those that allow surface atoms to have a higher number of neighbours:

**<center>surface atoms with more neighbours ∝ lower the energy of that surface type ∝ more likely to see that surface type</center>**

In this section, we will look at the energies of square and triangle surfaces. To do this we first need to calculate the energy related to only the surface, known as the surface energy ($E_{S}$). This requires us to calculate the energy of our metal in the bulk and subtract this from the energy of our surface.


<center>$E_{S} = \frac{E_{SM} - N_{SM} \times E_{BM}}{N_{SM}}$</center>


where $E_{SM}$ is the surface energy of your surface model (eV), $E_{BM}$ is the energy per atom of your bulk model (eV/atom), and $N_{SM}$ is the number of atoms in your surface model.

## Surface energy of the square surface

To begin, run the following python code below to get thre bulk energy of your material. This may take a few seconds or a minute to perform.

```python
from surface_energy_of_square_and_triangle import get_energy_of_bulk_model
element = 'Ag'
E_BM, lattice_constant = get_energy_of_bulk_model(element)
print('E_BM = ''{0:.3}'.format(E_BM),'eV / atom') 
```

Next, we will get the energy of your square surface model. The surface model will be shown when you run the following python code also.

```python
from surface_energy_of_square_and_triangle import get_energy_of_square_surface_model
E_SM_square, N_SM_square, slab_square = get_energy_of_square_surface_model(element,lattice_constant)
print('E_SM_square = ''{0:.3}'.format(E_SM_square),'eV')
print('N_SM_square = '+str(N_SM_square)+' atoms')
```

Now that we have $E_{SM}$, $N_{SM}$, and $E_{BM}$, now we will use the equation $E_{S} = \frac{E_{SM} - N_{SM} \times E_{BM}}{N_{SM}}$ to get the surface energy for our surface, E_{S}:

```python
E_S_square = 0.5 * ((E_SM_square - 6 * E_BM)/N_SM_square)
print('E_S_square = ''{0:.3}'.format(E_S_square),'eV / atom')
```

We have the surface energy for our square surface model, but the units are in eV/atoms. We would like this energy in J/m$^2$. To do this, just run the following python code below

```python
from surface_energy_of_square_and_triangle import eV_per_atom_into_J_per_m_squared
E_S_Jm2_square = eV_per_atom_into_J_per_m_squared(E_S_square,slab_square,N_SM_square)
print('E_S_Jm2_square = '+str(E_S_Jm2_square)+' J/m2')
```
