# 1.2: Using ASE


For these simulations, we will be using the Atomic Simulation Environment (ASE) program. ASE is a set of tools written in Python3 that allows us the construct and manipulate atomic structures. We can also use ASE to run simple calculations. 

In this project two main ASE modules will be used, the first being the ``ase.cluster`` module, which includes the octahedron and wulff construction scripts 




producing nanoparticle
models as shown in Figure 1, left. Another module that will be used is the ase.molecule
module; this allows for production of molecule models as shown in Figure 1, right.

```python
from ase.visualize import view
from ase.lattice.cubic import FaceCenteredCubic

element = 'Ag'
bulk = FaceCenteredCubic(size=(6,6,6),symbol=element,pbc=(1,1,1))

view(bulk)
```
