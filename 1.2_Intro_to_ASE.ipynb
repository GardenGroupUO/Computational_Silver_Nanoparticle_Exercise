## Having a play with ASE

To see how ASE works, we are going to use it to make a silver nanoparticle that we can then view. To do this, run the code below by pressing on the $\blacktriangleright$ play button. Hopefully, you will see a silver nanoparticle once it has run.

```python
from Computational_Silver_Nanoparticle_Exercise_Data.viewer import viewer
from ase.lattice.cubic import FaceCenteredCubic

element = 'Ag'
bulk = FaceCenteredCubic(size=(6,6,6),symbol=element,pbc=(1,1,1))

vr = viewer(bulk)
vr
```
