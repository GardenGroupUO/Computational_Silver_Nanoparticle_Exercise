## Having a play with ASE

To see how ASE works, we are going to use it to make a silver nanoparticle that we can then view. To do this, run the code below by pressing on the $\blacktriangleright$ play button. Hopefully, you will see a silver nanoparticle once it has run.

```python
from ase.cluster import wulff_construction

symbol = 'Ag'
surf_e_100 = 1.1808             
surf_e_111 = 1.0
esurf = [surf_e_100,surf_e_111]
surfaces = [(1,0,0),(1,1,1)]
size = 500

atoms = wulff_construction(symbol,surfaces,esurf,size,'fcc',rounding='above')

from ase.build import molecule

molecule_name = 'CH3CH2OCH3' #'H2O','NaCl','CH3CH2OCH3','CH3SiH3'
atoms = molecule(molecule_name)

from Computational_Silver_Nanoparticle_Exercise_Data.viewer.x3d_viewer import view_x3d

view_x3d(atoms)
```
