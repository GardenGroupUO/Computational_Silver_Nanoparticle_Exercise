# Part 2.5: How do Metallic Nanoparticles Differ?

Not all metals will look the same in nanoparticle form with roughly the same number of atoms. Different metal atoms have different surface energies ($E_S$) for their square and triangle surfaces surfaces. We are often interested in the ratio of the square surface energy:triangle surface energy, also called the (100):(111) energy ratio, where

* square surfaces are also called (100) surfaces
* triangle surfaces are also called (111) surfaces

In the following section, we will look at what the most stable nanoparticles containing about 500 atoms are for different types of metal elements. Here, the (100):(111) energy ratio for each metal is different. Note: We often set (111) to 1.0 as a convention, and scale the (100) number so that it is $\frac{(100)\:surface\:energy}{(111)\:surface\:energy}$
