```python
#from google.colab import output
#output.enable_custom_widget_manager()

from ase.build import molecule

molecule_name = 'CH3CH2OCH3' #'H2O','NaCl','CH3CH2OCH3','CH3SiH3'
atoms = molecule(molecule_name)

from ase.visualize import view

def viewer(atoms):
    v = view(atoms,viewer='ngl')
    v.custom_colors({'Mn':'green','As':'blue'})
    v.view._remote_call("setSize", target="Widget", args=["600px", "600px"])
    v.view.center()
    v.view.background='#ffc'
    v.view.parameters=dict(clipDist=-1)
    return v

v = viewer(atoms)
v
```
