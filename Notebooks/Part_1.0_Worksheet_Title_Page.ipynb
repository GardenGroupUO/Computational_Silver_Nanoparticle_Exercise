<h2> <font color="195EBC"><b>Introduction</b></font> </h2>

Nanoscience refers to the science of very small materials, between 1-100 nm. Nanoscience is particularly interesting because materials on the nanoscale show vastly different properties from the bulk. For instance, on the nanoscale, silver is no longer silver coloured but can be most of the colours of the rainbow!

<center>

</center>

An extremely nice experiment has been designed by a team at Wilfred Laurier University, Canada, to make silver nanoprisms that give the colours in the picture above. The experiment is described in [this link](https://pubs.acs.org/doi/10.1021/ed100166g) and you can watch a video of us performing it below.

<font color="red" size="">**insert video here**</font>

While this experiment is really cool, it is hard to understand exactly what the nanoprisms look like and why they give the colours they do. Therefore we have designed this computational exercise to help illustrate some of the principles of the nanoprism experiment, as well as nanoparticles more generally. We will learn In this online lesson, we will learn what nanoparticles are, what nanoparticles look like and what determines their shape, and why nanoparticles of different sizes have different colours.
