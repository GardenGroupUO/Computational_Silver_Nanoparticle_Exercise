## Analysing the reactions that happen

Lets analyse what was happening as we were preparing and creating our silver nanoprisms. 

1. Add sodium citrate, silver nitrate (AgNO<sub>3</sub>), and hydrogen peroxide (H<sub>2</sub>O<sub>2</sub>) to your solution.
2. Add the desired about of potassium bromide (KBr).
3. Add sodium borohydride (NaBH<sub>4</sub>).

So what is happening?

### What are sodium borohydride (NaBH<sub>4</sub>) and hydrogen peroxide (H<sub>2</sub>O<sub>2</sub>) used for?

To begin, silver in our AgNO<sub>3</sub> solution has a positive charge (Ag<sup>+</sup>). Ag<sup>+</sup> can not form nanoparticles becaue they are positively charge, meaning that one Ag<sup>+</sup> ion is electrically repelled from other Ag<sup>+</sup> ions. To make nanoparticles with Ag<sup>+</sup>, we need to reduce them so that they are not electrically charge. We do this using sodium borohydride (NaBH<sub>4</sub>), which is a very good reducing agent.

<center>
<font size=3>
NaBH<sub>4</sub> + 8AgNO<sub>3</sub> + 4H<sub>2</sub>O $\rightarrow$ Na[B(OH)<sub>4</sub>] + 8Ag + 8H<sup>+</sup> + 8NO<sub>3</sub><sup>-</sup>
</font>
</center>

This allows us to make Ag nanoparticles. However, if we used only NaBH<sub>4</sub> to reduce silver, it would reduce Ag<sup>+</sup> to Ag so quickly we would make a silver mirror instead of Ag nanoparticles. 

To reduce the speed that Ag<sup>+</sup> is reduced to Ag, we also include hydrogen peroxide (H<sub>2</sub>O<sub>2</sub>) to our solution which is an oxidising agent. This oxidises some of the Ag we have created back to Ag<sup>+</sup>.

<center>
<font size=3>
2Ag + H<sub>2</sub>O<sub>2</sub> + 2H<sup>+</sup> $\rightarrow$ 2Ag<sup>+</sup> + 2H<sub>2</sub>O
</font>
</center>

The volumes and concentrations of NaBH<sub>4</sub> (reducing agent) and H<sub>2</sub>O<sub>2</sub> (oxidising agent) that we have used in this experiment have been chosen so that we are creating Ag at the right speed to make Ag nanoparticles.

### What is sodium citrate used for?

The reduction of Ag<sup>+</sup> to Ag with NaBH<sub>4</sub> creates H<sup>+</sup> in the process. This H<sup>+</sup> is used by H<sub>2</sub>O<sub>2</sub> to oxidise Ag back to Ag<sup>+</sup>. However, if we have too much H<sup>+</sup>, H<sub>2</sub>O<sub>2</sub> will oxidise Ag back to Ag<sup>+</sup> too quickly! To remove the  H<sup>+</sup> that is created by NaBH<sub>4</sub> and Ag<sup>+</sup>, we used citrate that acts as a buffer.

<center>
<font size=3>
H<sup>+</sup> + citrate<sup>-</sup> $\rightarrow$ H-citrate
</font>
</center>

Citrate keeps the solution at a pH of around 6.4 and remove enough H<sup>+</sup> so that H<sub>2</sub>O<sub>2</sub> only oxidises Ag back to Ag<sup>+</sup> at the rate required for us to make Ag nanoprisms. 

The negative charge of citrate has also been hyprothesised to stabilise the nanoparticles as they are created. 

### What is potassium bromide (KBr) used for?



Br<sup>-</sup> ions are attracted and will bind to silver. The Br<sup>-</sup> ion attaches to the silver on the surface of the nanoprism. This prevents the Ag nanoprism from growing any further.
